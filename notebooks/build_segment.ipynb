```python
import pandas as pd
import numpy as np
```

```python
data_claster = pd.read_csv('../data/interim/claster_data.csv')
data = pd.read_csv('../data/interim/msk_2018_2021_clear.csv')
```

```python
data.head()
```

```python
data = data.merge(data_claster, left_on='Unnamed: 0', right_on='Unnamed: 0')
```

```python
data
```

```python
def drop_field(df : pd.DataFrame):
    df.drop('Unnamed: 0', axis=1, inplace=True)
    df.drop('X', axis=1, inplace=True)
    df.drop('Y', axis=1, inplace=True)
    df.drop('LAT', axis=1, inplace=True)
    df.drop('LON', axis=1, inplace=True)
    return df
```

```python
data = drop_field(data.copy())
```

```python
print(data["claster"].unique())
print(data["object_type"].unique())
print(data["rooms"].unique())
```

```python
data["segment"] = "k=" + data["claster"].astype(str) + "_" + "t=" + data["object_type"].astype(str) + "_" +"r="+ data["rooms"].astype(str)
print(data["segment"].unique())
len(data["segment"].unique())
```

```python
def del_field(df):
    df.drop('claster', axis=1, inplace=True)
    df.drop('object_type', axis=1, inplace=True)
    df.drop('rooms', axis=1, inplace=True)
    df.drop('timestamp', axis=1, inplace=True)
    return df
```

```python
data = del_field(data.copy())
```

```python
data.head()
```

```python
data_mean = data.groupby(['year', 'month', 'segment']).agg({'target' : ['mean']}).reset_index()
```

```python
data_mean.columns = [ 'year', 'month', 'segment', 'target']
```

```python
data_mean.head()
```

```python
data_mean['data'] = data_mean['year'].astype(str) + '-' + data_mean['month'].astype(str)
```

```python
data_mean.drop('year', axis=1, inplace=True)
data_mean.drop('month', axis=1, inplace=True)
```
